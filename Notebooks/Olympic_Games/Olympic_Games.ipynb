## Aggregating and Ranking

```python
olympics
```

```python
sum_vs_win = pd.crosstab(olympics.Country, olympics.Edition)
```

```python
sum_vs_win
```

```python
men_women =  pd.crosstab(olympics.Country, olympics.Gender)
men_women
```

```python
medals_by_cats = pd.concat([sum_vs_win, men_women], axis = 1)
medals_by_cats 
```

```python
medals_by_cats["Total"] = medals_by_cats.Summer + medals_by_cats.Winter
```

```python
medals_by_cats
```

```python
medals_by_cats.sort_values("Total", ascending = False, inplace = True)
```

```python
medals_by_cats
```

```python
ranks = medals_by_cats.rank(ascending = False, method = "average")
ranks
```

```python
top_25 = ranks.head(25) #We will be using top 25 countries that overall are the best
```

```python
plt.figure(figsize = (25, 10))
ax=plt.axes()
sns.heatmap(top_25.T,cmap='coolwarm',annot=True)
ax.set_title('Top 25 Countries by Medal Count')
#plt.savefig('Top_25_Countries_by_Medal_Count', bbox_inches='tight')
plt.show()
```

## Are countries equally successful in Summer and Winter olympic games?

```python
top_25.sort_values("Summer")
```

```python
top_25.sort_values("Winter")
```

```python
plt.figure(figsize = (25, 10))
ax=plt.axes()
sns.heatmap(top_25.sort_values(by = "Summer").T,cmap='coolwarm_r',annot=True,fmt='2.0f')
ax.set_title('Top 25 Countries in Summer by Medal Count')
#plt.savefig('Top_25_Countries_in_Summer_by_Medal_Count', bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize = (25, 10))
ax=plt.axes()
sns.heatmap(top_25.sort_values(by = "Winter").T,cmap='coolwarm',annot=True)
ax.set_title('Top 25 Countries in Winter by Medal Count')
#plt.savefig('Top_25_Countries_in_Winter_by_Medal_Count', bbox_inches='tight')
plt.show()
```

```python
top_25.Summer.sub(top_25.Winter).sort_values()
```

```python
rank_diff = top_25.Summer.sub(top_25.Winter).sort_values().to_frame()
rank_diff
```

```python
rank_diff.rename(columns = {0:"Difference"}, inplace = True)
```

```python
plt.figure(figsize = (25, 4))
ax=plt.axes()
sns.heatmap(rank_diff.T,cmap='bwr_r',annot=True, center = 0)
ax.set_title('Which country performs better in Summer(Red) vs Winter(Blue)')
#plt.savefig('Heat_Summer_Winter', bbox_inches='tight')
plt.show()
```
